## Cleaning individual rows

- let’s remove redundant data from the ‘Name’ column
- This involves removing punctuation, numbers, special characters, etc., and retaining only alphabets.

````python
def name_clean(text):
  return re.sub(r"[^a-zA-Z0-9 ]", "", text)
  df['name']=df['name'].apply(lambda x:name_clean(x))
````

````python
df['rate'].unique()
````

````python
import numpy as np

````

````python
df["rate"]=df["rate"].replace("NEW", np.nan)
df['rate']=df['rate'].replace('NaN',np.nan)
df['rate']=df['rate'].replace('-',np.nan)
df['rate']=df['rate'].replace('nan',np.nan)
````

````python
def remove_5(value: str):
    if type(value)==str:
        value_new=value.split('/')[0]
        return value_new
    return value
df['rate']=df['rate'].apply(remove_5)
#fun to remove /5
````

````python
df['rate']=df['rate'].astype(float)

print(df['rate'].dtypes)

#apply float func to data
````

Noticing the ‘cost’ attribute, we can observe that they are string values containing commas. Below is the code to convert them into numbers


````python
df=df.rename(columns={'approx_cost(for two people)':'Cost','listed_in(type)':'Type','listed_in(city)':'City','online_order':'Takes online orders?','book_table':'Has table booking?'})
````

````python
## function to remove commas and convert the values
## into numbers
def cost(value):
    value = str(value)
    if "," in value:
        value = float(value.replace(",",""))
        return value
    else:
        return float(value)
df['Cost'] = df['Cost'].apply(cost)
print(df['Cost'].head())
````

##What are the top 10 highest-rated restaurants?
##Often or not, many customers visit a place or choose a place to eat by looking at the restaurant’s ratings. So let us find out the top 10 restaurants that people visit.

````python
grp2=df.groupby('name')['rate'].mean().reset_index().sort_values(by='rate',ascending=False)[0:10]
print(grp2)

````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot
sns.scatterplot(data=grp2, x='name', y='rate', s=100, color='red')
for i, e in enumerate(grp2.rate):
    plt.text(i, e, round(e, 2), fontsize=15, horizontalalignment='center')
plt.xlabel('Restaurant Name', size=20)
plt.ylabel('Average rating out of 5', size=20)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20, rotation=90)
plt.show()
````

##What is the average price for two people based on the type of service?

````python
df.columns
````

````python


d = df.groupby('Type')['Cost'].mean().reset_index().sort_values(by='Cost')

plt.figure(figsize=(30, 10))
plt.plot(d['Type'], d['Cost'], 'o--r', ms=10)
plt.xlabel('Service type', size=20)
plt.ylabel('Average cost', size=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title('Cost based on the type of service provided', size=20)
for i, e in enumerate(d['Cost']):
    plt.text(i, e + 1, round(e, 1), fontsize=15, horizontalalignment='center')

plt.show()
````
